```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
x_values = np.linspace(-2, 2, 400)  # Bereich der x-Werte
k_values = np.arange(1, 101)  # Bereich der k-Werte

# Berechnung des Verhältnisses für jedes x
ratios = []

for x in x_values:
    ratio_list = []
    for k in k_values:
        # Berechnung der Terme
        a_k = (-2)**k * (k**2 + 2) / (k**3 + k) * x**(3 * k)
        a_k_plus_1 = (-2)**(k + 1) * ((k + 1)**2 + 2) / ((k + 1)**3 + (k + 1)) * x**(3 * (k + 1))
        
        # Berechnung des Verhältnisses
        ratio = abs(a_k_plus_1 / a_k)
        ratio_list.append(ratio)
    
    # Berechnung des mittleren Verhältnisses
    ratios.append(np.mean(ratio_list))

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, ratios, label='Mittleres Verhältnis |a_{k+1} / a_k|')
plt.axhline(1, color='red', linestyle='--', label='Konvergenzgrenze')
plt.xlabel('x')
plt.ylabel('Mittleres Verhältnis')
plt.title('Mittleres Verhältnis des Verhältnisses der Terme')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Beispiel Daten
n_span = np.linspace(0, 2 * np.pi, 100)
results = np.sin(n_span)

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=3)

def init():
    ax.set_xlim(0, 2 * np.pi)
    ax.set_ylim(-1, 1)
    return line,

def update(frame):
    line.set_data(n_span[:frame], results[:frame])
    return line,

ani = FuncAnimation(fig, update, frames=len(n_span), init_func=init, blit=True)
plt.show()
```
