```python
from sympy import symbols, diff, exp

# Definition der Variablen
r, a_B, e = symbols('r a_B e')

# Definiere das Potential phi(r)
phi_r = (e / r) * (1 + r / a_B) * exp(-2 * r / a_B)

# Erste Ableitung von phi(r) nach r
phi_prime_r = diff(phi_r, r)

# Berechnung der ersten Ableitung
phi_prime_r.simplify()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 1  # Radius of the wheel (arbitrary units)
omega = 1  # Angular velocity (rad/s)

# Time array
t = np.linspace(0, 4 * np.pi, 500)  # one full rotation

# x(t) and y(t) based on the provided equations
x_t = R * (omega * t - np.sin(omega * t))
y_t = R * (1 - np.cos(omega * t))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_t, y_t, label=r'Trajektorie $(x(t), y(t))$', color='blue')
plt.xlabel(r'$x(t)$')
plt.ylabel(r'$y(t)$')
plt.title(r'Trajektorie des Steinchens auf dem rollenden Reifen')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
