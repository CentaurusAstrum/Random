```python
import numpy as np 
import math
import matplotlib.pyplot as plt
```

```python
result = []
sum = 1

for i in range(1, 7000):
    sum *= (i + 2) / (i + 3)
    result.append(sum)

plt.plot(result)
plt.show()
print(result[-1])
```

```python
result = []
sum = 0

for i in range(1, 100):
    sum += (-10)**i / math.factorial(i)
    result.append(sum)

plt.plot(result)
plt.show()
print(result[-1])
```

```python
result = []

sum = 0

for i in range(2, 100):
    sum = np.log(1+ 10/i)
    result.append(sum)

plt.plot(result)
plt.show()
print(result[-1])
```
